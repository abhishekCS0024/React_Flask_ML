```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("fedesoriano/stroke-prediction-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import os
print(os.listdir(path))
```

```python
df = pd.read_csv(f"{path}/healthcare-dataset-stroke-data.csv")
```

```python
df.head()
```

```python
df.drop('id',axis='columns',inplace=True)
```

```python
df.info()
```

```python
df.dtypes
```

```python

def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}')
```

```python
df1=df
```

```python
print_unique_col_values(df1)
```

```python
# here we apply binary & one hot encoding for encoding of categorical data to numeric
df1['ever_married'] = df1['ever_married'].apply(lambda x: 1 if x == 'Yes' else 0)
df1['Residence_type'] = df1['Residence_type'].apply(lambda x: 1 if x == 'Urban' else 0)
```

```python
df2=df1
```

```python

# df2 = pd.get_dummies(data=df, columns=['gender','work_type','smoking_status'])
df2 = pd.get_dummies(data=df, columns=['gender','work_type','smoking_status'], dtype=int)

df2.columns
```

```python
for i in df2.columns:
  if(df2[i].dtypes!=bool):
    print(i)



```

```python
df2.dtypes
```

```python

cols_to_scale = ['age',

'avg_glucose_level',
'bmi',]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.dtypes
```

```python
df2.head(3)
```

```python
df2.isnull().sum()
```

```python
df2.shape
```

```python
df2['bmi']=df2['bmi'].fillna(df2['bmi']mean())
```

```python
df2.isnull().sum()
```

```python

X = df2.drop('stroke',axis='columns')
y = df2['stroke']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
df2.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(12,),activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python

y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```
